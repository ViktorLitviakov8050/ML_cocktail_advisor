```python
# First cell - imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Second cell - load data
df = pd.read_csv('data/cocktails.csv')
df.head()
```

```python

print("Dataset shape:", df.shape)
print("\nNumber of unique cocktails:", df['name'].nunique())
print("\nMost common glass types:")
print(df['glassType'].value_counts().head())



```

```python

lemon_drop = df[df['name'].str.contains('Aviation', case=False, na=False)]
print("\nAviation details:")
if len(lemon_drop) > 0:
    print("\nName:", lemon_drop['name'].values[0])
    print("Category:", lemon_drop['category'].values[0]) 
    print("Glass Type:", lemon_drop['glassType'].values[0])
    print("\nIngredients:")
    ingredients = eval(lemon_drop['ingredients'].values[0])
    measures = eval(lemon_drop['ingredientMeasures'].values[0])
    for ing, meas in zip(ingredients, measures):
        print(f"- {ing}: {meas}")
    print("\nInstructions:")
    print(lemon_drop['instructions'].values[0])
else:
    print("Classic Lemon Drop Martini not found in the dataset")






```
